```python
# Splitting the data into Features & Targets (Train & test split)

X = new_df.drop(columns='Class', axis=1)
y = new_df['Class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report ,confusion_matrix, auc, roc_auc_score, roc_curve

def classify(model):
    """
    Trains the given model on the dataset, calculates accuracy, and displays classification report and confusion matrix.

    Parameters:
        model: A scikit-learn model object

    Returns:
        None
    """    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    score = accuracy_score(y_test, y_pred)*100
    print("Here's your Accuracy(%):", round(score))
    print('\n')
    
    cls_report = classification_report(y_test, y_pred, output_dict=True)
    df = pd.DataFrame(cls_report).transpose()
    print("Classification Report:") 
    print(df)
    

    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='3.0f', cmap="summer")
    plt.title('And the confusion matrix', y=1.05, size=15)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    # Plotting ROC Curve
    y_probs = model.predict_proba(X_test)
    y_probs_positive = y_probs[:,1]
    
    # Calculate fpr, tpr and thresholds
    fpr, tpr, thresholds = roc_curve(y_test, y_probs_positive)
    plt.plot(fpr, tpr, color='green', label='ROC')
    # Plot line with no predictive power (baseline)
    plt.plot([0,1], [0,1], color="violet", linestyle="--", label="Guessing")
    
    # Customize the plot
    plt.xlabel("False positive rate (fpr)")
    plt.ylabel("True positive rate (tpr)")
    plt.title("Receiver Operating Characteristic (ROC) curve")
    plt.legend()
    plt.show()
    
```

## *Logistics Regression*

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model)
```

## *Decision Tree*

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
classify(model)
```

## *Random Forest*

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
classify(model)
```

## *Naive Bayes*

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
classify(model)
```
